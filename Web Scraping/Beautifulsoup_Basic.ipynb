```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
def make_soup(theurl):
    thepage = urllib.request.urlopen(theurl)
    soup = BeautifulSoup(thepage, 'html.parser')
    return soup
```

```python
soup = make_soup("http://codesolutionlab.com/")
```

```python
ok = soup.prettify

print(ok)
```

```python
for i in soup.findAll('link'):
    print(i.get('href'))
    
```

```python
for divs in soup.findAll('div', {'class':'portfolio-wrapper'}):
    for images in divs.findAll('img'):
        print(images.get('src'))
    
```

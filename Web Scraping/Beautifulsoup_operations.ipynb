```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/gp/bestsellers/computers/1375424031/ref=zg_bs_1375424031_pg_1?ie=UTF8&pg=1"
```

```python
opn = urllib.request.urlopen(url).read()
```

```python
soup = BeautifulSoup(opn)
```

```python
# print(soup)
```

```python
# Other way

def make_soup(theurl):
    data = urllib.request.urlopen(theurl)
    soup = BeautifulSoup(data, 'html.parser')
    return soup
```

```python
soup = make_soup(url)
```

```python
print(soup)
```

```python
for divs in soup.findAll('div',{"class":"a-section a-spacing-none aok-relative"}):
    for images in divs.findAll('img'):
        print(images.get('src'))
        print(images.get('alt'))
```

```python
for divs in soup.findAll('div',{"class":"a-section a-spacing-none aok-relative"}):
    for prices in divs.findAll('span',{"class":"a-size-base a-color-price"}):
#         print(prices)
        print(prices.text)
    
```

```python
for divs in soup.findAll('div',{"class":"a-section a-spacing-none aok-relative"}):
    for ratings in divs.findAll('span',{"class":"a-icon-alt"}):
        print(ratings.text)
```
